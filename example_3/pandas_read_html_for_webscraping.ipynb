#### Using pandas read_html to Web Scrape Data for Data Science

#### Method 1
#### We could modularize the URL and year strings of text that may be useful for programmatic access (in the subsequent code cell)

```python
# The Building Blocks
year = '2019'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

# Combining the URL + year strings together
url = url_link.format(year)
url
```

#### Now, let's programmatically return a list of URL given a list of years (e.g. 2015,2016,2017,2018,2019).

```python
years = [2015,2016,2017,2018,2019]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

for year in years:
  url = url_link.format(year)
  print(url)
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
```

#### Read HTML webpage into pandas

```python
import pandas as pd
```

```python
# Using pd.read_html() function to read HTML webpage
# Данные загружаются в этот фрейм данных
df = pd.read_html(url, header = 0)
print(type(df))
print(df)
```

#### How many tables are there in the webpage? Важно определить сколько таблиц и с какой работаем

```python
len(df)
```

#### Select the next table - .т.к. на сайте одна таблица мы будем использовать 0, который    представляет 1 табл

#### Data cleaning - delete rows with headers

#### We can see that the table header is presented multiple times in several rows
#### Когда мы увидим слово age - мы будем удалять всю строку, поле age содержит номер

```python
df2019[df2019.Age == 'Age']
```

```python
# всего 26 полей которые содержат age
len(df2019[df2019.Age == 'Age'])
```

```python
df2019.shape
```

```python
# .drop() - функция которая поможет нам удалить все поля age из фрейма
#           проверка 734 - 708 = 26 строк удалены
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
```

#### Quick Exploratory Data Analysis - Быстрый исследовательский анализ данных

```python
import seaborn as sns
```

#### Making the histogram

```python
# df.PTS - переменная для построения точек
# kde=False - будет ложным, т.к. мы хотим сохранить исходную частоту, если 
#             мы хотим иметь фактическую частоту количество счетчиков, которое для каждого бара 
#             сколько игроков иметь - столько очков,
#             так что между 0 и 1 одно очко, сколько игроков есть около 40 с чем-то игроков имеют 1 очко
#             колличество игроков имеющих 35 или более очков будет меньше 10
#             у большинства очки от 0 до 20, и очень немногие имеют точки более 20

sns.distplot(df.PTS, kde=False)
```

```python
# kde=True - это будет вероятность не частота 
sns.distplot(df.PTS, kde=True)
```

#### Change bar line color

```python
sns.distplot(df.PTS, 
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2))
```

#### Change bar fill colors

```python
sns.distplot(df.PTS,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```
